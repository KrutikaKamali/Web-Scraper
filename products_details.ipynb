## Aim : To extract data from any website using Beautiful Soup

### Importing required libraries and initializing beautiful soup

```python
#Importing required libraries

import bs4
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
#Initializing Beautiful soup

url = 'https://www.flipkart.com/search?q=mobile+phone+under+30000&sid=tyy%2C4io&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_18_sc_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_18_sc_na_na&as-pos=1&as-type=RECENT&suggestionId=mobile+phone+under+10000%7CMobiles&requestId=a381be32-5a8e-4e09-b272-e2f6990b562f&as-backfill=on'

r = requests.get(url)
soup = BeautifulSoup(r.text,'lxml')
data = soup.findAll
```

#### Creating list of product names

```python
#Creating list of product names
Name = soup.findAll('div',class_ = '_4rR01T')
Name_list = []
for n in Name:
    n = n.text
    Name_list.append(n)
    
print(Name_list)
```

```python
len(Name_list)
```

#### Creating list of ratings

```python
#Creating list of ratings
Ratings = soup.findAll('div',class_ = '_3LWZlK')
Ratings_list = []
for r in Ratings:
    r = r.text
    Ratings_list.append(r)
    
del Ratings_list[24:]
print(Ratings_list)
```

```python
len(Ratings_list)
```

#### Creating list of descriptions

```python
#Creating list of descriptions
Desc = soup.findAll('ul',class_ = '_1xgFaf')
Desc_list = []
for de in Desc:
    de = de.text
    Desc_list.append(de)
print(Desc_list)
```

```python
len(soup.findAll('ul',class_ = '_1xgFaf'))
```

#### Creating list of price of products

```python
#Creating list of price of products
price = soup.findAll('div',class_ = '_30jeq3 _1_WHN1')
Price_list = []
for pr in price:
    pr = pr.text
    Price_list.append(pr)

Price_list
```

```python
len(soup.findAll('div', class_ = '_30jeq3 _1_WHN1'))
```

#### Creating list of original price of products

```python
#Creating list of original price of products
Opr = soup.findAll('div', class_ = '_3I9_wc _27UcVY')
Ori_price_list = []
for opr in Opr:
    opr = opr.text
    Ori_price_list.append(opr)
print(Ori_price_list)
```

```python
len(soup.findAll('div',class_ = '_3I9_wc _27UcVY'))
```

#### Creating list of offer percent

```python
#Creating list of offer percent
offper = soup.findAll('div',class_ = '_3Ay6Sb')
Offer_per = []
for op in offper:
    op = op.text
    Offer_per.append(op)
print(Offer_per)    
```

#### Iterating through pages

```python
#Iterating through pages

Name_list = []
Ratings_list = []
Desc_list = []
Price_list = []
Ori_price_list = []
Offer_per = []

for i in range(1,3):
    url = ("https://www.flipkart.com/search?q=mobile+phone+under+30000&sid=tyy%2C4io&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_18_sc_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_18_sc_na_na&as-pos=1&as-type=RECENT&suggestionId=mobile+phone+under+10000%7CMobiles&requestId=a381be32-5a8e-4e09-b272-e2f6990b562f&as-backfill=on&page="+str(i))        
    r = requests.get(url)
    print(r)
    
    soup = BeautifulSoup(r.text,'lxml')
    Name = soup.findAll('div',class_ = '_4rR01T')
    for n in Name:
        n = n.text
        Name_list.append(n)
    print("Name:", Name_list)
    print("Name length:", len(Name_list))
    
    Ratings = soup.findAll('div',class_ = '_3LWZlK')
    for r in Ratings:
        r = r.text
        Ratings_list.append(r)
    del Ratings_list[len(Name_list):]    
    print("Ratings:", Ratings_list)
    print("Ratings length:", len(Ratings_list))
    
    Desc = soup.findAll('ul',class_ = '_1xgFaf')
    for de in Desc:
        de = de.text
        Desc_list.append(de)
    print("Desc:", Desc_list)
    print("Desc length:", len(Desc_list))
    
    Price = soup.findAll('div',class_ = '_30jeq3 _1_WHN1')
    for pr in Price:
        pr = pr.text
        Price_list.append(pr)
    print("Price:", Price_list)
    print("Price length:", len(Price_list))
    
    Opr = soup.findAll('div', class_ = '_3I9_wc _27UcVY')
    for opr in Opr:
        opr = opr.text
        Ori_price_list.append(opr)
    
    if len(Ori_price_list)<len(Name_list):
        Ori_price_list.append("0")
    elif len(Ori_price_list)>len(Name_list):
        del Ori_price_list[len(Name_list):] 
    
    print("Opr:", Ori_price_list)
    print("Opr length:", len(Ori_price_list))
    
    Offper = soup.findAll('div',class_ = '_3Ay6Sb')
    for op in Offper:
        op = op.text
        Offer_per.append(op)
    del Offer_per[len(Name_list):]
    print("Offper:", Offer_per)    
    print("Offper length:", len(Offer_per))
    
dict = {'Name':Name_list,'Rating':Ratings_list,'Description':Desc_list,'Price':Price_list,'Original_price':Ori_price_list,
        'Offer_percentage':Offer_per}
df = pd.DataFrame(dict).shift()[1:]
df
```
