## Aim : To extract data from any website which has Load more button

### Importing required libraries

```python
#Importing required libraries & creating User agent

import requests
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36'}
```

```python
# Setting the base url

url = 'https://smarthistory.org/americas-before-1900/'
response = requests.get(url,headers= headers)
print(response.status_code)
```

```python
# Checking status_code
(response.status_code)
```

```python
# Checking status_code for the link in the load more button
url = 'https://smarthistory.org/wp-json/smthstapi/v1/objects?tag=938&page=1'
response = requests.get(url,headers= headers)
print(response.status_code)
```

```python
#Iterating through the links in the load more button

url = 'https://smarthistory.org/wp-json/smthstapi/v1/objects?tag=938&page={}'
pg_num = 1
title =[]
while True:
    response = requests.get(url.format(pg_num),headers= headers) #using format method to concat int to str      
    data = response.json() #grabbing all data from response and storing as json
    d = data['posts']      #entering dict and 'posts'
    for i in d:            #iterating through dicts
        for key, values in i.items():
            if key == 'title':
                title.append(values.strip())
    if data.get('remaining') and int(data.get('remaining')) >0: #checking for the remaining no. of links
        pg_num+= 1
    else:
        break

#printing the o/p data        
for i in title:
    print(i)
```

#### Converting into DataFrame

```python
#Converting into dataframe
import pandas as pd
df = pd.DataFrame(title)
df
```

#### Converting Dataframe to a CSV file

```python
df.to_csv('json_scraper')
```
